```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None  # default='warn'
from pathlib import Path
from IPython.display import display
```

## TEST
### read 

```python
all_test = pd.read_csv("Data/result_matrix/all_test.csv",";", na_values='.', low_memory=False) #Read participant-file to a dataframe
display(all_test)

mean_all = all_test['total_mean'].mean()
mean_std_all = all_test['total_mean'].std()

avg_emp_before = all_test['empathy_before'].mean()
avg_emp_after = all_test['empathy_after'].mean()
avg_emp_diff = all_test['empathy_diff'].mean()

print("mean_all")
print(mean_all)
print("mean_std_all")
print(mean_std_all)
print("avg_emp_before")
print(avg_emp_before)
print("avg_emp_after")
print(avg_emp_after)
print("avg_emp_diff")
print(avg_emp_diff)
```

## CONTROL
### read 

```python
all_control = pd.read_csv("Data/result_matrix/all_control.csv",";", na_values='.', low_memory=False) #Read participant-file to a dataframe
display(all_control)


mean_all = all_control['total_mean'].mean()
mean_std_all = all_control['total_mean'].std()

avg_emp_before = all_control['empathy_before'].mean()
avg_emp_after = all_control['empathy_after'].mean()
avg_emp_diff = all_control['empathy_diff'].mean()

print("mean_all")
print(mean_all)
print("mean_std_all")
print(mean_std_all)
print("avg_emp_before")
print(avg_emp_before)
print("avg_emp_after")
print(avg_emp_after)
print("avg_emp_diff")
print(avg_emp_diff)
```

## TEST
### scatter
- empathy before vs mean
- empathy after vs mean
- empathy deaf vs mean 

```python
font = {'size':22}
plt.rc('font', **font)

plt.figure(figsize=(10,10))
plt.scatter(all_test['empathy_before'], all_test['total_mean'])
plt.xlabel("Empathy score before")
plt.ylabel("Avg (all attempts) pupil size (mm)")

coef = np.corrcoef(all_test['empathy_before'], all_test['total_mean'])
print("Correlation coefficient: empathy_before vs total_mean" )
print(coef)

plt.figure(figsize=(10,10))
plt.scatter(all_test['empathy_after'], all_test['total_mean'])
plt.xlabel("Empathy score after")
plt.ylabel("Avg (all attempts) pupil size (mm)")

coef2 = np.corrcoef(all_test['empathy_after'], all_test['total_mean'])
print("Correlation coefficient: empathy_after vs total_mean" )
print(coef2)


plt.figure(figsize=(10,10))
plt.scatter(all_test['empathy_diff'], all_test['total_mean'])
plt.xlabel("Empathy score after - before")
plt.ylabel("Avg (all attempts) pupil size (mm)")

coef3 = np.corrcoef(all_test['empathy_diff'], all_test['total_mean'])
print("Correlation coefficient: empathy_diff vs total_mean" )
print(coef3)

plt.figure(figsize=(10,10))
plt.scatter(all_test['empathy_mean'], all_test['total_mean'])
plt.xlabel("Avg empathy score")
plt.ylabel("Avg (all attempts) pupil size (mm)")

coef4 = np.corrcoef(all_test['empathy_mean'], all_test['total_mean'])
print("Correlation coefficient: empathy_mean vs total_mean" )
print(coef4)

#plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

## TEST
### plot
- attempt mean with std d, fraction saccades and fixations

```python
font = {'size':22}
plt.rc('font', **font)
from matplotlib.backends.backend_pdf import PdfPages

pp = PdfPages('all_test.pdf')

i = 0
for key, group in all_test.groupby('participant'):
    i = i+1
    #fig, ax = plt.subplots(figsize=(15,10))
    fig=plt.figure(figsize=(30,10))

    plt.subplot(1, 2, 1)
    plt.xlabel("attempt")
    plt.ylabel("avg_diameter (mm)")
    plt.title('Mean and std, participant ' + str(key))
    plt.errorbar(group['attempt'], group['attempt_mean'], group['attempt_std_d'], linestyle='-', marker='o', capsize=5, ecolor="green", elinewidth = 0.5)

    # plt.subplot(1, 2, 2)
    # plt.xlabel("attempt")
    # plt.ylabel("% of all events")
    # plt.title('Fraction: saccades and fixations, participant ' + str(key))
    # plt.plot(group['attempt'], group['fraction_sacc'], label = "saccades")
    # plt.plot(group['attempt'], group['fraction_fix'],label = "fixations")
    # plt.legend( loc='lower right', borderaxespad=3.5)
    plt.savefig('result_test/' + str(i) + '.png')
    #pp.savefig(fig)
    #group.plot('attempt', 'mean', yerr='std_d', label = 'participant ' + str(key), ax=ax)
#pp.close()
```
